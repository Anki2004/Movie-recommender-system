## Text based recommendation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming merged is your DataFrame containing movie data
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_metrics = vectorizer.fit_transform(merged['genres'].values.astype('U'))

def recommend_movies(input_text, user_rating, num_recommendations=5):
    input_vector = vectorizer.transform([input_text])
    similarities = cosine_similarity(input_vector, tfidf_metrics)

    # Using argsort to get indices of movies sorted by similarity
    sorted_indices = similarities.argsort()[0][-num_recommendations:][::-1]

    recommended_movies = []
    for index in sorted_indices:
        recommended_movies.append(merged.iloc[index])

    return recommended_movies

user_rating = 5.0
recommended_movies = recommend_movies("horror | comedy", user_rating, num_recommendations=5)

print("Recommended movies are:")
for movie in recommended_movies:
    print('Name:', movie['title'])
    print('Genre:', movie['genres'])
    print('Rating:', movie['rating'])
    print('Time Stamp:', movie['timestamp'])
    print('---')
```
